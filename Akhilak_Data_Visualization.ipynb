### Task 4
Create the plot by adding two axes to a figure object at `[0,0,1,1]` and `[0.2,0.5,0.4,0.4]`. use x,y, and z arrays to recreate the plot below. Notice the x limits and y limits on the inner plot

```python
fig4 = plt.figure()
axis_1 = fig4.add_axes([0,0,1,1])
axis_2 = fig4.add_axes([0.2,0.5,0.4,0.4])
x = np.arange(0,100,1)
y = 2*x
z = x**2

axis_1.plot(x,z)
axis_2.plot(x,y)
axis_1.set_xlabel('x')
axis_1.set_ylabel('z')
axis_2.set_xlabel('x')
axis_2.set_ylabel('y')

axis_1.grid(False)
axis_2.grid(False)

axis_2.set_title("Zoom In")
axis_2.set_xlim(20,25)
axis_2.set_ylim(30,60)
```

### Task 5

### Task 10
Load the csv file 2014_World_Power_Consumption and display its first five rows

```python
wpc_df = pd.read_csv(r'2014_World_Power_Consumption', sep=',')
wpc_df.head()
```

### Task 11
Create a Choropleth Plot of the Power Consumption for Countries using the data and layout dictionary.

```python
fig = px.choropleth(wpc_df, locations = 'Country', color = 'Power Consumption KWH', 
                    locationmode = 'country names',                   
                    color_continuous_scale = "ylorrd",
                    scope = "world", labels = {'Power Consumption KWH':'Power Consumption KWH'}, 
                    hover_name = 'Country', title = 'Power Consumption in KWH')

fig.update_layout(margin = {"r":0,"t":0,"l":0,"b":0})
fig.update_geos(projection_type="natural earth")
pyo.iplot(fig)
```

### Task 13
Create a plot that displays the Voting-Age Population (VAP) per state

```python
fig = px.choropleth(election_df, locations = 'State Abv', color = 'Voting-Age Population (VAP)', 
                    locationmode = 'USA-states',                   
                    color_continuous_scale = "ylorrd",
                    scope = "usa", labels = {'Voting-Age Population(VAP)':'Voting-Age Population(VAP)'}, 
                    hover_name = 'State',
                    title = 'Voting-Age Population')

fig.update_geos(scope="usa", showlakes = True, lakecolor = 'rgb(85, 173, 240)') 
fig.update_layout(margin = {"r":0,"t":0,"l":0,"b":0})
pyo.iplot(fig)
```
